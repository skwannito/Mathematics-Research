```python
# Sommes suffixes : suffix_sum[i] = somme(primes_numbers[i:])
suffix_sum = [0] * (len(primes_numbers) + 1)
for i in range(len(primes_numbers) - 1, -1, -1):
    suffix_sum[i] = suffix_sum[i + 1] + primes_numbers[i]

# Algo pour chercher la première décomposition en somme de nombre premier

def research(np, limite):
    solution = []

    def rec(i, reste):
        if reste == 0:
            return solution.copy()

        if reste < 0:
            return None

        if i >= limite:
            return None

        # Élagage : même en prenant tous les nombres restants,
        # on ne pourra jamais atteindre "reste".
        if suffix_sum[i] - suffix_sum[limite] < reste:
            return None

        # On prend le nombre
        solution.append(primes_numbers[i])
        r = rec(i + 1, reste - primes_numbers[i])
        solution.pop()

        if r is not None:
            return r

        # On ne le prend pas
        return rec(i + 1, reste)

    return rec(0, np)

```

```python
def est_somme_parfaite(t):
    return t == list(primes_numbers[:len(t)])
```

```python
# 

for i, np in enumerate(primes_numbers):
    if np > 11:
        t = research(np, i)
        print(f"{np:04}", t, len(t))
```
